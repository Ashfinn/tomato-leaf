```python
import matplotlib.pyplot as plt
import cv2
import os
import random

dataset_dir = "dataset"  # your folder path
class_folders = sorted(os.listdir(dataset_dir))  # get class folder names

# just for nice labeling
label_names = [name.replace("Tomato__", "").replace("_", " ") for name in class_folders]

fig, axs = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axs.flat):
    class_path = os.path.join(dataset_dir, class_folders[i])
    image_file = random.choice(os.listdir(class_path))
    img_path = os.path.join(class_path, image_file)

    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.imshow(img)
    ax.set_title(label_names[i], fontsize=9)
    ax.axis('off')

plt.tight_layout()
plt.savefig("dataset_samples.png", dpi=300)
plt.show()
```
